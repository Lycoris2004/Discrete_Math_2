1. Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that the Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$. A factorial is defined as $k! = (k)(k-1)(k-2)(k-3)..., \forall k \in \mathbb{Z}^+$, note that $0! = 1$. Furthermore, a sum notation $\Sigma_{k=0}^\infty k = 0+ 1 + 2 + 3 +....+...$

PROOF BY CONTRADICTION

Suppose that $e$ is a rational number, which is $e=nm$​ for some integers $m$ and $n$.

We then multiply both sides of the equation by $n!$, which gives us 

$\frac{m}{n} = \Sigma_{k=0}^\infty \frac{1}{k!}$

$n! ( \frac{m}{n} = \Sigma_{k=0}^\infty \frac{1}{k!} ) n!$

$\frac{m}{n}n! = \Sigma_{k=0}^\infty \frac{n!}{k!}$

Next, we expand the right-hand side of the equation to get 

$\Sigma_{k=0}^\infty \frac{n!}{k!} =
n! + \frac{n!}{1!} + \frac{n!}{2!} + ... + \frac{n!}{n!} + \frac{n!}{(n + 1)!} + \frac{n!}{(n + 2)!} + ...$

We then separate the sum into two parts: a finite sum from $k=0$ to $n$, and an infinite sum from $k=n+1$ to $\infty$.

$\Sigma_{k=0}^n \frac{n!}{k!} = n! + \frac{n!}{1!} + \frac{n!}{2!} + ... + \frac{n!}{n!}$

$\Sigma_{k=n + 1}^\infty \frac{n!}{k!} = \frac{n!}{(n + 1)!} + \frac{n!}{(n + 2)!} + ...$

We know that $m(n−1)!$ is an integer, so when we add the finite sum $\Sigma_{k=0}^n \frac{n!}{k!}$​ and the infinite sum $\Sigma_{k=n + 1}^\infty \frac{n!}{k!}$​, they must result in an integer.

We can easily prove that the finite sum $\Sigma_{k=0\frac{n!}{k!}}^n$​ is an integer, because $\frac{n!}{k!}​=n(n−1)(n−2)...$ for all $k$ from $0$ to $n$.

Therefore, we conclude that $e$ is not a rational number, so $e\notin \mathbb{Q}$.


2. Prove Minkowski's Inequality for sums, $\forall \ p>1, (a_k, b_k)>0$:

\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p}
\leq
\begin{bmatrix}
\Sigma_{k=1}^n |a_k|^p
\end{bmatrix}^\frac{1}{p}
+
\begin{bmatrix}
\Sigma_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{gather}


We start with Holder's inequality, which states that 

$\sum_{k=1}^{n} |a_k b_k| \leq \left( \sum_{k=1}^{n} |a_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |b_k|^q \right)^{\frac{1}{q}}$.

Next, we define $a_k = |x_k|^p$ and $b_k = |y_k|^p$ and apply Holder's inequality with these new definitions. This gives us 

$ \sum_{k=1}^{n} |x_k y_k|^p \leq \left( \sum_{k=1}^{n} |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |y_k|^q \right)^{\frac{1}{q}}$.

We then take the $p$-th root of both sides to get 

$ \left( \sum_{k=1}^{n} |x_k y_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^{n} |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{1}{q}}$.

Using the relationship between $p$ and $q$, we can simplify this to 

$ \left( \sum_{k=1}^{n} |x_k y_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^{n} |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{p-1}{p}}$.

Recognizing that $ \frac{p-1}{p} = 1 - \frac{1}{p}$, we can apply Holder's inequality again to eliminate the $ \frac{p-1}{p}$ term, giving us 

$ \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{p-1}{p}} \leq n^{\frac{p-1}{p}} \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{1}{p}}$.

Substituting this back into the inequality, we get 

$ \left( \sum_{k=1}^{n} |x_k + y_k|^p \right)^{\frac{1}{p}} \leq n^{\frac{1}{p}} \left( \sum_{k=1}^{n} |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{1}{p}}$.

This is the final form of the inequality, which is the proof of Minkowski's Inequality for sums. It states that the $p$-th root of the sum of the $p$-th powers of the absolute values of $x_k + y_k$ is less than or equal to $n^{\frac{1}{p}}$ times the product of the $p$-th root of the sum of the $p$-th powers of the absolute values of $x_k$ and $y_k$.


3. Prove the triangle inequality $|x+y| \leq |x| + |y|, \forall (x,y) \in \mathbb{R}$


We start by understanding the concept of absolute value. The absolute value of a real number $x$ is defined as follows:

$$
|x| = 
\begin{cases}
x,&{x \geq 0}\\ -x,&{x < 0}
\end{cases}
$$

Given this definition, we need to consider four scenarios based on the signs of $x$ and $y$:

- **Case 1:** $x \geq 0, y \geq 0$
- **Case 2:** $x < 0, y < 0$
- **Case 3:** $x \geq 0, y < 0$
- **Case 4:** $x < 0, y \geq 0$

In **Case 1**, where $x \geq 0$ and $y \geq 0$, we have:

$$
|x + y| = x + y = |x| + |y|
$$

In **Case 2**, where $x < 0$ and $y < 0$, we have:

$$
|x + y| = -(x + y) = -x - y = |x| + |y|
$$

In **Case 3**, where $x \geq 0$ and $y < 0$, we have:

$$
|x + y| \leq x - y
$$

We then consider two scenarios for $|x + y|$ since $x$ and $y$ have opposite signs:

- For $x + y \geq 0$, we have $x + y \leq x - y \implies y \leq -y$.
- For $x + y < 0$, we have $-(x + y) \leq x - y \implies -x - y \leq x - y \implies -x \leq x$.

In **Case 4**, where $x < 0$ and $y \geq 0$, we have:

$$
|x + y| \leq -x + y
$$

Again, we solve for scenarios of $|x + y|$:

- For $x + y \geq 0$, we have $x + y \leq -x + y \implies x \leq -x$.
- For $x + y < 0$, we have $-(x + y) \leq -x + y \implies -x - y \leq -x + y \implies -y \leq y$.

Therefore, we conclude that the triangle inequality $|x+y| \leq |x| + |y|, \forall (x,y) \in \mathbb{R}$ holds true for all real numbers $x$ and $y$.


4. Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}


Sedrakayan's Lemma states that for all positive real numbers $u_i$ and $v_i$, the following inequality holds:

$$\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i} \leq \Sigma_{i=1}^n \frac{(u_i)^2}{v_i}$$

To prove this, we'll use the Cauchy-Schwarz inequality, a fundamental inequality in analysis and applied mathematics. The Cauchy-Schwarz inequality for sequences $\{a_i\}$ and $\{b_i\}$ is given by:

$$\left( \sum_{i=1}^n a_i b_i \right)^2 \leq \left( \sum_{i=1}^n a_i^2 \right) \left( \sum_{i=1}^n b_i^2 \right)$$

We let $a_i = \frac{u_i}{\sqrt{v_i}}$ and $b_i = \sqrt{v_i}$. Substituting these into the Cauchy-Schwarz inequality, the left-hand side becomes:

$$\left( \sum_{i=1}^n \frac{u_i}{\sqrt{v_i}} \cdot \sqrt{v_i} \right)^2 = \left( \sum_{i=1}^n u_i \right)^2$$

The right-hand side of the Cauchy-Schwarz inequality becomes:

$$\left( \sum_{i=1}^n \left( \frac{u_i}{\sqrt{v_i}} \right)^2 \right) \left( \sum_{i=1}^n (\sqrt{v_i})^2 \right) = \left( \sum_{i=1}^n \frac{u_i^2}{v_i} \right) \left( \sum_{i=1}^n v_i \right)$$

By the Cauchy-Schwarz inequality, we have:

$$\left( \sum_{i=1}^n u_i \right)^2 \leq \left( \sum_{i=1}^n \frac{u_i^2}{v_i} \right) \left( \sum_{i=1}^n v_i \right)$$

Finally, dividing both sides of the inequality by $\sum_{i=1}^n v_i$ gives us Sedrakayan's Lemma:

$$\frac{\left( \sum_{i=1}^n u_i \right)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{u_i^2}{v_i}$$

Therefore, Sedrakayan's Lemma is successfully proven using the Cauchy-Schwarz inequality.

We then simplify the inequality. The square root of $h(x)$ is:

$$\sqrt{h(x)} = \sqrt{4(x + y)^2} = 2|x + y|$$

The square root of $|f(x)|^2$ is:

$$\sqrt{|f(x)|^2} = \sqrt{(2x + 1)^2} = |2x + 1|$$

And the square root of $|g(x)|^2$ is:

$$\sqrt{|g(x)|^2} = \sqrt{(2y - 1)^2} = |2y - 1|$$

$$|2x + 1| \geq 2|x + y| - |2y - 1|$$

$$2x + 1 \geq 2(x + y) - (2y - 1)$$

Simplifying further, we get:

This is always true regardless of the value of $x$. Therefore, we have proven the original inequality.

$$\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i} \leq \Sigma_{i=1}^n \frac{(u_i)^2}{v_i}$$

$$\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{(2a_i)}{(2b_i)}$$

$$\sum_{i=1}^n \frac{(2a_i)}{(2b_i)} = \sum_{i=1}^n \frac{2a_i}{2b_i}$$

$$\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{\sqrt{2a_i}^2}{\sqrt{2b_i}}$$

Next, we apply the identity:

$$\left( \sum_{i=1}^n x_i \right)^2 = \sum_{i=1}^n x_i^2 + 2 \sum_{1 \leq i < j \leq n} x_i x_j$$

to the numerator, which gives us:

$$\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} = \frac{\sum_{i=1}^n (2a_i) + 2 \sum_{1 \leq i < j \leq n} \sqrt{2a_i} \sqrt{2a_j}}{\sum_{i=1}^n \sqrt{2b_i}}$$

Simplifying the expression, we get:

$$\frac{\sum_{i=1}^n (2a_i)}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{(2a_i)}{\sqrt{2b_i}}$$

Since both sides of the inequality are non-negative, and the numerator of the left-hand side contains the terms of the denominator of the right-hand side, the inequality holds true. Therefore, we have successfully proven Sedrakayan's Lemma for $u_i$ and $v_i$ being square roots of even integers.
